# Feature importance for Appliances
<br>The code asses the importance of each feature in the code (minute, hour, day, week, month) in predicting the usage the usage of it, across all the houses the appliance is present in. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score
import datetime as datetime
```

```python

def breakdown(a):
    global u
    u = pd.DataFrame(index=app.index, columns=cols) #usage table
    u.set_index(pd.DatetimeIndex(u.index),inplace=True)
    u.fillna(0,inplace=True)
    u['month']=u.index.month
    u['week']=u.index.week
    u['day']=u.index.dayofweek
    u['hour']=u.index.hour
    u['minute']=u.index.minute
    u[a][(app[a]>30)]=1
```

```python
# The data is split in 75% traning 25% testing datasets 
def RTF_feat(a,i):

    #     Under Sample the data 

    ones = len(u[a][(u[a] == 1)])
    zero_indices = u[u[a] == 0].index
    random_indices = np.random.choice(zero_indices,ones, replace=False)
    ones = u[u[a] == 1].index
    under_sample_indices = np.concatenate([ones,random_indices])
    under_sample = u.loc[under_sample_indices]
    
    global f
    f = pd.DataFrame() # place to save the feature importance
    

    #     Fit model to the data
    cols = ['month','week','hour','minute','day']
    X = under_sample[cols]
    y = under_sample[a]

    if(X.empty):                    # if after undersampling the the traning data becomes empty, the fitting operation will return an error, thus it shouldbe avoided
        print ("Not enough data")
    else:
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
        clf_rf = RandomForestClassifier(n_estimators=25, random_state=12)
        clf_rf.fit(X_train, y_train)
        y_pred_class = clf_rf.predict(X_test)
        f[i]= clf_rf.feature_importances_

```
